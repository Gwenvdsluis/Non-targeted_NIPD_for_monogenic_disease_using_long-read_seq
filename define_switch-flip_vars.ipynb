```python
import pandas as pd
import numpy as np
```

```python
wf_hvar = pd.read_csv("/hpc/umc_laat/gvandersluis/data/Ont_data_nhung/HG002/wf_hvar_ROI/rn_wf_hvar.wf_snp.vcf.gz", sep="\t", comment='#', names=['CHROM','POS','ID','REF','ALT','QUAL','FILTER','INFO','FORMAT', "sample"])
df_splitted = wf_hvar['sample'].str.split(":")
wf_hvar["GT"] = df_splitted.str[0]
wf_hvar["GQ"] = df_splitted.str[1]
wf_hvar["DP"] = df_splitted.str[2]
wf_hvar["PS"] = df_splitted.str[5]


```

```python
wf_hvar["Switch_relevance"] = "None"
wf_hvar
```

```python
loci_switch = pd.read_csv("/hpc/umc_laat/gvandersluis/data/Ont_data_nhung/HG002/wf_hvar_ROI/switches_ROI.bed", sep="\t", names=["CHROM", "START", "END", "ID"])
loci_switch.head(40)#[loci_switch["START"]==26622424]
```

```python

HB_ID = loci_switch["ID"].unique()
HB_ID.tolist()
```

```python
## TO do:
# add column to the variant dataframe: switch relevance ++
# per row in the switch bed file ++
# If 1 switch has occured in a haploblock; given locations are called switch and the ones in between switch associated ++
# If multiple: figure out if they are flips or switches and call them accordingly

## Options:
# Switch
# Switch associated
# Flip
# Flip associated

## create new vcf with the extra label
hb_id = HB_ID.tolist()
# Loop through the unique HBs
for id_tag in hb_id:
    if len(loci_switch[loci_switch["ID"] == id_tag]) == 1:
        chrom = loci_switch[loci_switch["ID"] == id_tag]["CHROM"].iloc[0]
        start = loci_switch[loci_switch["ID"] == id_tag]["START"].iloc[0].astype(int)
        end = loci_switch[loci_switch["ID"] == id_tag]["END"].iloc[0].astype(int)
        wf_hvar.loc[(wf_hvar["CHROM"] == chrom) & (wf_hvar["POS"].isin([start, end])), "Switch_relevance"] = "Switch"
        wf_hvar.loc[(wf_hvar["CHROM"] == chrom) & (wf_hvar["POS"].astype(int) > start) & (wf_hvar["POS"].astype(int) < end), "Switch_relevance"] = "Switch Associated"
    else:
        subset = loci_switch[loci_switch["ID"] == id_tag].reset_index(drop=True)
        previous_flip = "No" # to indicate if the previous switch was part of flip
        for i in range(len(subset)-1):
            chrom_1 = subset.loc[i]["CHROM"]
            start_1 = subset.loc[i]["START"].astype(int)
            end_1 = subset.loc[i]["END"].astype(int)

            chrom_2 = subset.loc[i+1]["CHROM"]
            start_2 = subset.loc[i+1]["START"].astype(int)
            end_2 = subset.loc[i+1]["END"].astype(int)
            if end_1 == start_2 and previous_flip == "No":
                wf_hvar.loc[(wf_hvar["CHROM"] == chrom_1) & (wf_hvar["POS"].astype(int) == end_1), "Switch_relevance"] = "Flip"
                wf_hvar.loc[(wf_hvar["CHROM"] == chrom_1) & (wf_hvar["POS"].astype(int) >= start_1) & (wf_hvar["POS"].astype(int) < end_1), "Switch_relevance"] = "Flip Associated"
                wf_hvar.loc[(wf_hvar["CHROM"] == chrom_2) & (wf_hvar["POS"].astype(int) > start_2) & (wf_hvar["POS"].astype(int) <= end_2), "Switch_relevance"] = "Flip Associated"
                previous_flip = "Yes"
            elif end_1 == start_2 and previous_flip == "Yes":
                wf_hvar.loc[(wf_hvar["CHROM"] == chrom_1) & (wf_hvar["POS"].astype(int) == end_1), "Switch_relevance"] = "Flip"
                wf_hvar.loc[(wf_hvar["CHROM"] == chrom_2) & (wf_hvar["POS"].astype(int) > start_2) & (wf_hvar["POS"].astype(int) <= end_2), "Switch_relevance"] = "Flip Associated"
                previous_flip = "Yes"
            elif end_1 != start_2 and previous_flip == "No":
                wf_hvar.loc[(wf_hvar["CHROM"] == chrom_1) & (wf_hvar["POS"].isin([start_1, end_1])), "Switch_relevance"] = "Switch Flank"
                wf_hvar.loc[(wf_hvar["CHROM"] == chrom_1) & (wf_hvar["POS"].astype(int) > start_1) & (wf_hvar["POS"].astype(int) < end_1), "Switch_relevance"] = "Switch Associated"
                previous_flip = "No"
            elif previous_flip == "No" and i == len(subset)-1:
                wf_hvar.loc[(wf_hvar["CHROM"] == chrom_1) & (wf_hvar["POS"].isin([start_1, end_1])), "Switch_relevance"] = "Switch Flank"
                wf_hvar.loc[(wf_hvar["CHROM"] == chrom_1) & (wf_hvar["POS"].astype(int) > start_1) & (wf_hvar["POS"].astype(int) < end_1), "Switch_relevance"] = "Switch Associated"
                wf_hvar.loc[(wf_hvar["CHROM"] == chrom_2) & (wf_hvar["POS"].isin([start_2, end_2])), "Switch_relevance"] = "Switch Flank"
                wf_hvar.loc[(wf_hvar["CHROM"] == chrom_2) & (wf_hvar["POS"].astype(int) > start_2) & (wf_hvar["POS"].astype(int) < end_2), "Switch_relevance"] = "Switch Associated"

          
```

```python
wf_hvar.to_csv("/hpc/umc_laat/gvandersluis/data/Ont_data_nhung/HG002/wf_hvar_ROI/switch_association.csv", index=False)
```
