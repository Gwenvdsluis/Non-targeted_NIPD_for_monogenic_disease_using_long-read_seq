```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
genes = pd.read_csv("/hpc/umc_laat/gvandersluis/data/Ref_HG/BIOMART_nc_genes.txt.gz", sep = "\t")
```

```python
genes["gene_length"] = genes["Gene end (bp)"] - genes["Gene start (bp)"]
genes
```

```python

plt.hist(genes["gene_length"], bins=50)
plt.xlabel("Gene length")
plt.ylabel("Count")
plt.title("Distribution of gene lengths")
plt.show()
```
